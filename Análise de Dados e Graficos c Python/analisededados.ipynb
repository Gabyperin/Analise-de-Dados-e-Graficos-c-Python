```python
# faturamento por loja
dados.groupby(["loja", "cidade"]).preco.sum()
```

```python
# media de faturamentos de cada loja das cidades
dados.groupby(["estado","loja", "cidade"]).preco.mean()
```

```python
# .to_excel("Media de Faturamento.xlsx") -> cria uma tabela no excel 
dados.groupby(["estado","loja", "cidade"]).preco.mean().to_excel("Media de Faturamento.xlsx")
```

```python
import plotly_express as px
```

```python
px.histogram(dados, x="loja", y="preco",title= "Vendas por Lojas", color="forma_pagamento", text_auto= True)
```

```python
# vamos armazenar esse grafico em uma variavel 
grafico = px.histogram(dados, x="loja", y="preco",title= "Vendas por Lojas", color="forma_pagamento", text_auto= True)
```

```python
# criando um arquivo html para exportar 
grafico.write_html("Gráfico Vendas por Lojas.html")
```

## Criando Múltiplos Gráficos

```python
colunas = ["loja", "cidade", "estado", "tamanho", "local_consumo"]
```

```python
for coluna in colunas:
    grafico = px.histogram(dados, 
                           x=coluna,
                           y="preco",
                           title= f"Faturamento por {coluna} ", 
                           color="forma_pagamento", 
                           text_auto= True)
    grafico.write_html(f"Gráfico {coluna}.html")
```
